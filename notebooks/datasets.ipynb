```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
```

```python
def print_dataset_info(X: np.ndarray) -> None:
    print("- Shape:")
    print(f"    - {X.shape[0]} patrones.")
    print(f"    - {X.shape[1]} características.")
```

MNIST
---

```python
mnist_bunch = fetch_openml("mnist_784")
mnist_data = mnist_bunch["data"].to_numpy(dtype=np.float32)
mnist_labels = mnist_bunch["target"].to_numpy(dtype=np.int32)
print_dataset_info(mnist_data)
```

```python
fig, axs = plt.subplots(2, 5, figsize=(15, 6))

class_idx = 0
for row in range(len(axs)):
    for col in range(len(axs[0, :])):
        axs[row, col].axis("off")
        axs[row, col].set_title(f"{class_idx}")

        indices = np.where(mnist_labels == class_idx)[0]
        rand_idx = np.random.choice(indices)
        img = mnist_data[rand_idx, :].reshape(28, 28)
        axs[row, col].imshow(img, cmap="gray")
        
        class_idx += 1

plt.show()
```

CIFAR-10
---

```python
cifar_bunch = fetch_openml("cifar_10_small")
cifar_data = cifar_bunch["data"].to_numpy(dtype=np.float32)
cifar_labels = cifar_bunch["target"].to_numpy(dtype=np.int32)
print_dataset_info(cifar_data)
```

```python
fig, axs = plt.subplots(2, 5, figsize=(15, 6))

class_idx = 0
for row in range(len(axs)):
    for col in range(len(axs[0, :])):
        axs[row, col].axis("off")
        axs[row, col].set_title(f"{class_idx}")

        indices = np.where(cifar_labels == class_idx)[0]
        rand_idx = np.random.choice(indices)
        img = cifar_data[rand_idx, :].reshape((3, 32, 32)).transpose(1, 2, 0)
        img /= 255.0
        axs[row, col].imshow(img)
        
        class_idx += 1

plt.show()
```
